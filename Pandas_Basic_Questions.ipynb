# How do you merge two DataFrames in Pandas?


```python
# https://pandas.pydata.org/docs/user_guide/merging.html
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html
```

1) the merge method is more versatile and allows us to specify columns beside the index to join on for both dataframes
2) Merge joins are faster and uses less memory than hash joins
3) The merge() method updates the content of two DataFrame by merging them together, using the specified method(s)

**merge() for combining data on common columns or indices**

```python
pd.merge(
    left,
    right,
    how="inner",
    on=None,
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
```

```python
*left   --> Required  # A DataFrame, a Series to merge with
*right  --> Required  # A DataFrame, a Series to merge with
*how  -->'left'
         'right'
         'outer'
         'inner'
         'cross'    # Optional. Default 'inner'. Specifies how to merge
*on  --> String
         List       # Optional. Specifies in what level to do the merging
*left_on --> String
             List   # Optional. Specifies in what level to do the merging on the DataFrame to the left
*right_on --> String
              List   # Optional. Specifies in what level to do the merging on the DataFrame to the right
*left_index  --> True
                 False  # Optional. Default False. Whether to use the index from the left DataFrame as join key or not
*right_index --> True
                 False  # Optional. Default False. Whether to use the index from the right DataFrame as join key or not
*sort --> True
          False     # Optional. Default False. Specifies whether to sort the DataFrame by the join key or not
*suffixes --> List  # Optional. Default '_x', '_y''. Specifies a list of strings to add for overlapping columns
*copy --> True
          False  # Optional. Default True. Specifies whether to keep copies or not
*indicator --> True
               False
               String  # Optional. Default False. Specifies whether to add a column in the DataFrame with information about the source of each row
* validate --> String  # Optional. Checks if the mergin is of a specified type
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
```

```python
df1 ,"  ", df2
```

```python
pd.merge(df1,df2, left_on='lkey', right_on='rkey')           ### very important
#df1.merge(df2, left_on='lkey', right_on='rkey')
```

# How do you join two DataFrames in Pandas?


```python
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html
```

1) The join() method inserts column(s) from another DataFrame, or Series.
2) This method does not change the original DataFrame.
3) Parameters on, lsuffix, and rsuffix are not supported when passing a list of DataFrame objects.

**join() for combining data on a key column or an index**

```python
DataFrame.join(other, on=None, how='left', lsuffix='', rsuffix='', sort=False, validate=None)
```

```python
*other --> Required  # A DataFrame, a Series or a list of DataFrames.
*on --> String
        List   # Optional. Specifies in what level to do the joining
*how --> 'left'
         'right'
         'outer'
         'inner'    # Optional. Default 'left'. Specifies which index to use
*lsuffix --> Sring  # Optional. Default '', Specifies a string to add for overlapping columns
*rsuffix --> Sring  # Optional. Default '', Specifies a string to add for overlapping columns
*sort --> True
          False     # Optional. Default False. Specifies whether to sort the DataFrame by the join key or not
```

# How do you handle missing values in a DataFrame?


--> many ways are there depends upon usage the type of handling been used 

```python
1) Drop rows or columns that have a missing value     # df.dropna(axis=1)
2) Drop rows or columns that only have missing values # df.dropna(how="all")
3) Drop rows or columns based on a threshold values   # df.dropna(thresh=4)
4) Drop based on a particular subset of columns       # df.dropna(subset=["col 1", "col 2"])
5) Fill with a constant value                         
6)* Fill with an aggregated value   (# most usage)    # df["col 1"].fillna(df["col 1"].mean())
7) Replace with the previous or next value            # df.fillna(method="bfill"
8) Fill by using another data frame #(when both same) # df1.fillna(df2)
```

    --> Polynomial Interpolation
    --> Interpolation Through Padding

# How do you create a pivot table in Pandas?


1) Pivot table is used to summarize data in which includes various statistical concepts

```python
df = pd.DataFrame({'Name': ['John', 'Sammy', 'Stephan', 'Joe', 'Emily', 'Tom'],
                   'Gender': ['Male', 'Female', 'Male',
                              'Female', 'Female', 'Male'],
                   'Age': [45, 6, 4, 36, 12, 43]})
```

```python
df
```

```python
def age_bucket(age):
    if age <= 18:
        return "<18"
    else:
        return ">18"               ## # categorizing in age groups
```

```python
df['Age Group'] = df['Age'].apply(age_bucket)
```

```python
gender = pd.DataFrame(df.Gender.value_counts(normalize=True)*100).reset_index()
gender.columns = ['Gender', '%Gender']   ## calculating gender percentage
df = pd.merge(left=df, right=gender, how='inner', on=['Gender'])
```

```python
df
```

```python
table = pd.pivot_table(df, index=['Gender', '%Gender', 'Age Group'],  
                       values=['Name'], aggfunc={'Name': 'count',})    ## creating pivot table
```

```python
print(table)
```
